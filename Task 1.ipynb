## Objective:
Understand how to load, explore, summarize, and visualize a simple dataset using Pandas, Matplotlib, and Seaborn.


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

## Scatter Plot
Analyze relationship between Sepal Length and Petal Length.


```python
plt.figure()
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

## Histogram
Examine distribution of Sepal Length.


```python
plt.figure()
plt.hist(df["sepal_length"], bins=10)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Box Plot
Detect outliers and spread of Petal Length by Species.


```python
plt.figure()
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length by Species")
plt.show()
```

## Conclusion
